```python
import random
def generate_shares(secret, num_parties=3):
    shares = []
    for i in range(num_parties - 1):
        shares.append(random.randint(1, 100))
    shares.append(secret - sum(shares))
    return shares
# Private salaries
ahmed_salary = 70
ali_salary = 80
rashida_salary = 70
# Generate shares
ahmed_shares = generate_shares(ahmed_salary)
ali_shares = generate_shares(ali_salary)
rashida_shares = generate_shares(rashida_salary)
print("Ahmed shares:", ahmed_shares)
print("Ali shares:", ali_shares)
print("Rashida shares:", rashida_shares)
# Each person sums the shares they receive
sum_ahmed = ahmed_shares[0] + ali_shares[0] + rashida_shares[0]
sum_ali = ahmed_shares[1] + ali_shares[1] + rashida_shares[1]
sum_rashida = ahmed_shares[2] + ali_shares[2] + rashida_shares[2]
print("\nPartial sums:")
print("Ahmed's sum:", sum_ahmed)
print("Ali's sum:", sum_ali)
print("Rashida's sum:", sum_rashida)
# Global sum
global_sum = sum_ahmed + sum_ali + sum_rashida
print("\nGlobal sum (total salary):", global_sum)
```
